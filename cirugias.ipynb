```python
#@title 1. Importar librerias
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
#@title 2. Cargar Datos
data= pd.read_csv('/content/CirugiasMarzo.csv', encoding='utf-8', sep=';')
```

```python
#@title 3. Visualizar metadatos
data.info()
data.describe()
#data2.rename({'fecha ':'tiempo')
```

```python
data
```

```python
# Eliminar duplicados basados en la columna 'cedula'
data = data.drop_duplicates(subset=['NUMERO IDE'], keep='first')
```

```python
data.info()
data.describe()
```

```python
data['CÓDIGO MUNICIPIO '] = data['CÓDIGO DEPARTAMENTO '].astype(str) + data['CÓDIGO MUNICIPIO '].astype(str)
data['CÓDIGO MUNICIPIO '] = data['CÓDIGO MUNICIPIO '].str.zfill(5)
```

```python
data.head(3)
```

```python
data.insert(loc=1,column='TIPO REGISTRO', value="3")
data.insert(loc=2,column='CONSECUTIVO', value="")
data.insert(loc=3,column='REALIZO', value="1")
data.insert(loc=4,column='NOREALIZO', value="")
data.insert(loc=5,column='SEPROGRAMO', value="2")
data['CONSECUTIVO'] = range(1, len(data) + 1)
```

```python
data.info()
data.describe()
```

```python

# Crear un subconjunto de datos con algunas columnas
cirugias = data[
    ['TIPO REGISTRO',
     'CONSECUTIVO',
     'TIPO IDE.',
     'NUMERO IDE',
     'FECHA NACIMIENTO',
     'SEXO',
     'PRIMER APELLIDO PACIENTE',
     'SEGUNDO APELLIDO PACIENTE',
     'PRIMER NOMBRE PACIENTE',
     'SEGUNDO NOMBRE PACIENTE',
     'ENTIDAD ADMINISTRADORA',
     'CÓDIGO MUNICIPIO ',
     'CÓDIGO QX',
     'FECHA INGRESO',
     'FECHA PRO',
     'REALIZO',
     'NOREALIZO',
     'SEPROGRAMO',
    ]]
```

```python
# prompt: necesito sacar los paciente con nombre que empueza por c

# Filtrar pacientes cuyo primer nombre comienza con "C"
pacientes_c = cirugias[cirugias['PRIMER NOMBRE PACIENTE'].str.startswith('C', na=False)]

# Mostrar los resultados
pacientes_c
```

```python
cirugias.head(3)
```

```python
# prompt: agrupar por sexo

# Agrupar por sexo y contar la cantidad de registros
grouped_data = cirugias.groupby('SEXO').size().reset_index(name='Cantidad')

# Imprimir el resultado
grouped_data
```

```python

# Convertir 'FECHAs' to datetime objects
cirugias['FECHA NACIMIENTO'] = pd.to_datetime(cirugias['FECHA NACIMIENTO'], format='%d/%m/%Y', errors='coerce')
cirugias['FECHA INGRESO'] = pd.to_datetime(cirugias['FECHA INGRESO'], format='%d/%m/%Y', errors='coerce')
cirugias['FECHA PRO'] = pd.to_datetime(cirugias['FECHA PRO'], format='%d/%m/%Y', errors='coerce')
cirugias['CÓDIGO QX'] = pd.to_numeric(cirugias['CÓDIGO QX'], errors='coerce').astype('Int64')

```

```python
cirugias.head(3)
```

```python

# Guardar el DataFrame como archivo CSV
cirugias.to_csv('cirugia_new.csv', index=False)
```
